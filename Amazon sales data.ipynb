# Extract New Columns from Dataset

```python
df['Year of order']=pd.DatetimeIndex(df['Order Date']).year          #here we extract the year of the order from order date column..
df['Quarter of order']=pd.DatetimeIndex(df['Order Date']).quarter    #extract quarter of the order
df['Month of order']=pd.DatetimeIndex(df['Order Date']).month        #extract month of the order 
```

```python
df.head()            #our new column are here in the dataset...
```

```python
#we want to see our new column after the order date so let's do it...

df1=df[['Region','Country','Item Type','Sales Channel','Order Priority','Order Date','Year of order','Quarter of order','Month of order','Order ID','Ship Date','Units Sold','Unit Price','Unit Cost','Total Revenue','Total Cost','Total Profit']]
```

```python
df1                     #here is our modify dataframe
```
